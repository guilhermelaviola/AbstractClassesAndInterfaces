# **Tangent Line, Velocity & Acceleration**
A **tangent line** is a straight line that touches a curve at a single point, representing the curve's instantaneous direction or slope at that exact spot. In physics, this mathematical concept directly defines **velocity**, which is the rate of change of an object's position with respect to time, geometrically represented as the slope of the tangent line on a position-time graph. Furthermore, **acceleration** measures how quickly that velocity changes over time, serving as the slope of the tangent line on a velocity-time graph, meaning both physical concepts rely on the tangent line to describe precise, moment-to-moment motion.

## **Example: Tangent Line, Velocity & Acceleration**
The following example uses numpy and matplotlib that models a moving object, calculates its velocity and acceleration, and plots the tangent lines to visualize these concepts.

```python
# Importing all the necessary libraries and resources:
import numpy as np
import matplotlib.pyplot as plt

# Defining time and a position function: s(t) = t^3 - 3t^2 + 2t:
t = np.linspace(0, 3, 100)
position = t**3 - 3*t**2 + 2*t

# Calculate Velocity (v = ds/dt) and Acceleration (a = dv/dt) using derivatives:
velocity = 3*t**2 - 6*t + 2
acceleration = 6*t - 6

# Picking a specific point in time to find the tangent lines (e.g., t = 2 seconds):
t_match = 2.0
pos_at_t = t_match**3 - 3*t_match**2 + 2*t_match  # s(2) = 0
vel_at_t = 3*t_match**2 - 6*t_match + 2          # v(2) = 2 (Slope of position tangent)
acc_at_t = 6*t_match - 6                         # a(2) = 6 (Slope of velocity tangent)

# Defining the tangent line equations: y = m(x - x1) + y1:
tangent_pos = vel_at_t * (t - t_match) + pos_at_t
tangent_vel = acc_at_t * (t - t_match) + vel_at_t

# Plotting the results:
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(8, 12), sharex=True)

# Position Graph & Tangent Line (Slope = Velocity):
ax1.plot(t, position, label='Position $s(t)$', color='blue', lw=2)
ax1.plot(t, tangent_pos, '--', label=f'Tangent at $t={t_match}$ (Slope = Velocity = {vel_at_t})', color='cyan')
ax1.scatter([t_match], [pos_at_t], color='black', zorder=5)
ax1.set_ylabel('Position (m)')
ax1.set_title('Tangent Lines Visualizing Velocity and Acceleration')
ax1.legend()
ax1.grid(True)

# Velocity Graph & Tangent Line (Slope = Acceleration):
ax2.plot(t, velocity, label='Velocity $v(t)$', color='green', lw=2)
ax2.plot(t, tangent_vel, '--', label=f'Tangent at $t={t_match}$ (Slope = Accel = {acc_at_t})', color='lime')
ax2.scatter([t_match], [vel_at_t], color='black', zorder=5)
ax2.set_ylabel('Velocity (m/s)')
ax2.legend()
ax2.grid(True)

# Acceleration Graph:
ax3.plot(t, acceleration, label='Acceleration $a(t)$', color='red', lw=2)
ax3.scatter([t_match], [acc_at_t], color='black', zorder=5)
ax3.set_ylabel('Acceleration ($m/s^2$)')
ax3.set_xlabel('Time (s)')
ax3.legend()
ax3.grid(True)

plt.tight_layout()
plt.show()
```
