# **Applied Differential Mathematics**
Applied Differential Mathematics is the branch of mathematics that uses derivatives, differential equations, and related concepts to model, analyze, and solve real-world problems involving change, rates, and dynamic systems. It focuses on applying calculus tools to understand how quantities evolve over time or space in fields such as physics, engineering, biology, economics, data science, and environmental modeling.

```python
# Importing all the necessary libraries and resources:
import numpy as np
import matplotlib.pyplot as plt
import math
import sympy as sp
import scipy.signal as sig
```

```python
# Modeling temperature change (Newton's Law of Cooling):
k = 0.1
T_ambient = 25
T0 = 90
t = np.linspace(0, 60, 200)

T = T_ambient + (T0 - T_ambient) * np.exp(-k * t)

plt.plot(t, T)
plt.title('Cooling of Coffee Over Time')
plt.xlabel('Minutes')
plt.ylabel('Temperature (Â°C)')
plt.show()
```

## **Number Sets, Inequalities & Absolute Value**

```python
# Filtering valid sensor values:
sensor_values = [3.4, -2, 10, 5, -20]

# Keeping only values within [-5, 10]:
valid = [x for x in sensor_values if -5 <= x <= 10]

# Absolute deviation from target = 0:
abs_dev = [abs(x) for x in sensor_values]

valid, abs_dev
```

## **Functions: Definition, Graph, Operations & Common Types**

```python
# Demand function (Linear)
def demand(p):
    return 100 - 2*p

# Cost function (Quadratic):
def cost(q):
    return 0.05*q**2 + 5*q

p = np.linspace(0,50,200)
q = np.linspace(0,100,200)

plt.plot(p, demand(p), label='Demand')
plt.plot(q, cost(q), label='Cost')
plt.legend()
plt.show()
```

## **Important Functions: Even, Odd, Injective, Surjective & Bijective**

```python
# Checking if a function is injective on a domain:
def f(x): return 3*x + 2

domain = range(-5, 6)
is_injective = len(set(f(x) for x in domain)) == len(list(domain))
is_injective
```

```python
# Even/odd test:
def is_even(f, x): return f(x) == f(-x)
def is_odd(f, x): return f(x) == -f(-x)

g = lambda x: x**2 # Even
h = lambda x: x**3 # Odd

is_even(g, 3), is_odd(h, 3)
```

## **Limits & Lateral Limits**

```python
# Traffic flow approaching road capacity:
def flow(density):
    return density * (1 - density/100)  # Max at Density=50

left_limit = flow(50 - 1e-6)
right_limit = flow(50 + 1e-6)
left_limit, right_limit
```

## **Asymptotes**

```python
# Rational function modeling diminishing marginal utility:
x = np.linspace(1,100,200)
y = 100/x + 2  # Horizontal Asymptote at y = 2

plt.plot(x,y)
plt.axhline(2, color='red', linestyle='--')
plt.show()
```

## **Continuity & Intermediate Value Theorem**

```python
# Temperature sensor must cross zero:
temps = [-5, -2, 1, 4]

# If f(a) < 0 < f(b), it must cross 0
crosses_zero = any(temps[i] < 0 < temps[i+1] for i in range(len(temps)-1))
crosses_zero
```

## **Tangent Line, Velocity & Acceleration**

```python
# Car motion (position, velocity, acceleration):
t = sp.symbols('t')
s = 5*t**2 + 3*t  # Position

velocity = sp.diff(s, t)
acceleration = sp.diff(velocity, t)

velocity, acceleration
```

## **Differentiation Rules & Composite/Inverse**

```python
# Chain rule in cooling + fan system:
t = sp.symbols('t')
T = sp.exp(-0.1*t) # Temperature
fan_speed = sp.sqrt(T) # Depends on temperature

sp.diff(fan_speed, t)
```

## **Elementary Derivatives: Exponential, Logarithmic, Trigonometric & Hyperbolic**

```python
# Growth rate of bacteria (exponential derivative):
t = sp.symbols('t')
P = sp.exp(0.3*t)
sp.diff(P, t)
```

## **Applied Derivatives: Implicit Differentiation**

```python
# Circle expanding (related rates):
x, y = sp.symbols('x y', real=True)
r = sp.symbols('r', real=True)
expr = x**2 + y**2 - r**2

dy_dx = -sp.diff(expr, x) / sp.diff(expr, y)
dy_dx
```

## **Maxima & Minima**

```python
# Profit maximization:
q = sp.symbols('q')
revenue = 50*q - 0.3*q**2
profit = revenue - (10*q)

dprofit = sp.diff(profit, q)
critical_points = sp.solve(dprofit, q)
critical_points
```

## **Rolle's Theorem & Mean Value Theorem**

## **Extrema, Concavity & Inflection Points**

```python
# Detecting turning points in stock prices:
prices = np.array([30,32,35,33,31,34,36,35])
peaks = sig.find_peaks(prices)[0]
valleys = sig.find_peaks(-prices)[0]

peaks, valleys
```
